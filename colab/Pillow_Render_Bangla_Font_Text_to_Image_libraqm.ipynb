```
from PIL import ImageFont, ImageDraw, Image
import numpy as np
import matplotlib.pyplot as plt


image = np.zeros((200,600,3),np.uint8)

image = Image.fromarray(np.uint8(image * 255))


draw = ImageDraw.Draw(image)

# use a truetype font
#font = ImageFont.truetype("/content/BalooDa2-Medium.ttf", 24, layout_engine=ImageFont.LAYOUT_RAQM)
font = ImageFont.truetype("/content/BalooDa2-Medium.ttf", 24)


draw.text((20, 25), "আমাদের দেশের ক্রিকেটাররা এখন টেস্ট খেলায় জিততে জানে", font=font)
draw.text((20, 50), "গরমে সুস্থ থাকুন", font=font)


plt.figure(figsize = (20,20))
plt.imshow(image)
```
